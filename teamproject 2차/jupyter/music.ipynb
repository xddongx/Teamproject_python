```python


def emotion(e):
    if e == '불안한':
        emo = emotions[0]
    elif e == '화난':
        emo = emotions[1]
    elif e == '역겨운':
        emo = emotions[2]
    elif e == '행복한':
        emo = emotions[3]
    elif e == '슬픈':
        emo = emotions[5]
    elif e == '놀란':
        emo = emotions[6]
    else :
        emo = emotions[4]
    
    emotion_ = music[music.emotion == emo]
    print(e)
    rand_list = emotion_.sample()
    rand_list.emotion = e
    print(rand_list)
    return rand_list
```

```python
text = '무표정'
emos = emotion(text).values
```
