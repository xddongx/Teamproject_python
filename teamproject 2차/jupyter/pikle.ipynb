```python
import cv2
import face_recognition
import pickle
import tensorflow as tf
tf.reset_default_graph()

dataset_paths = ['emotion/anger/', 'emotion/happy/', 'emotion/sad/', 'emotion/surprised/']
names = ['anger', 'happy', 'sad', 'suprised']
number_images = 10
image_type = '.jpg'
encoding_file = 'encodings_face_rgb.pickle'
# cnn 또는 hog입니다. CNN 방법이 더 정확하지만 느립니다. HOG는 빠르지 만 정확도는 떨어집니다
model_method = 'cnn'

# 알려진 인코딩 및 알려진 이름의 목록을 초기화합니다.
knownEncodings = []
knownNames = []



```

```python
# 이미지 경로를 반복
for (i, dataset_path) in enumerate(dataset_paths):
    
    # 이름에서 사람 이름을 추출
    name = names[i]
    print(dataset_path)

    for idx in range(number_images):
        file_name = dataset_path + str(idx+1) + image_type

        # 입력 이미지를로드하고 BGR (OpenCV 주문)에서 dlib 주문 (RGB)으로 변환
        image = cv2.imread(file_name)
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        

        # 입력 이미지에서 각면에 해당하는 경계 상자의 (x, y) 좌표를 감지합니다.
        boxes = face_recognition.face_locations(rgb, model=model_method)
        

        # 얼굴의 얼굴 임베딩 계산
        encodings = face_recognition.face_landmarks(rgb, boxes)

        # 인코딩을 반복
        for encoding in encodings:
            # add each encoding + name to our set of known names and
            # encodings
            print(file_name, name, encoding)
            knownEncodings.append(encoding)
            knownNames.append(name)
        
# 얼굴 인코딩 + 이름을 디스크에 저장
data = {"encodings": knownEncodings, "names": knownNames}
f = open(encoding_file, "wb")
f.write(pickle.dumps(data))
f.close()
```

# 파일 이름변경

```python
import os 
def changeName(path):
    i = 1
    for filename in os.listdir(path):
        print(path+filename, '=>', path+str(i)+'.jpg')
        os.rename(path+filename, path+str(i)+'.jpg')
        i += 1
changeName('emotion/anger/')
```

```python
changeName('emotion/happy/')
```

```python
changeName('emotion/sad/')
```
